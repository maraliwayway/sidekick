```python
import pandas as pd 
data1 = pd.read_csv("vitamin_info_updated.csv")
data1
```

```python
data2 = pd.read_csv("vitamins_info.csv")
```

```python
# df.rename(columns={'old_column_name': 'new_column_name'}, inplace=True)
data2.rename(columns={'Indigenous Remedies': 'Natural Remedies'}, inplace=True)
# df.drop('column_name', axis=1, inplace=True)
data2.drop("Food Sources",axis=1, inplace=True) 
data2
```

```python
# merged_df = pd.merge(df1, df2, on='common_column')
merged_datas = pd.merge(data1, data2, on='Vitamin')
merged_datas
# 1,000 IU of vitamin D is equivalent to 25 mcg
```

```python
# getting rid of some extra strings in the natural remedies section
remove_words = ["of", "diet:", "include", "remedies:"]
pattern = r'^(?:' + '|'.join(remove_words) + r')\s*'
merged_datas['Natural Remedies'] = merged_datas['Natural Remedies'].str.replace(pattern, '', regex=True)
merged_datas
```

```python
merged_datas.to_csv("vitamin_final.csv", index=False)
```
